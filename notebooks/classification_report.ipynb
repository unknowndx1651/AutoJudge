### Classification

```python
# import os;
# os.chdir("..")

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix

from src.classifier import xtrain_final,xtest_final,ytrain,ytest
```

### 1.logistic regression

```python
LR=LogisticRegression(random_state=42,solver="lbfgs",class_weight="balanced",max_iter=8000)
LR.fit(xtrain_final, ytrain)

ypred1=LR.predict(xtest_final);
```

```python
print("Logistic Regression accuracy: ", LR.score(xtest_final,ytest))
print(confusion_matrix(ytest, ypred1))
print("Logistic Regression f1 score: ", f1_score(ytest, ypred1, average="macro"))
```

### 2. Linear Support Vector Classifier (SVC)

```python
svc=SVC(random_state=42, class_weight="balanced",max_iter=8000, C=1)
svc.fit(xtrain_final, ytrain)

ypred2=svc.predict(xtest_final);
```

```python
print("SVC accuracy: ", svc.score(xtest_final,ytest))
print(confusion_matrix(ytest, ypred2))
print("SVC f1 score: ", f1_score(ytest, ypred2, average="macro"))
```

### 3. Random Forest Classifier:

```python
RF=RandomForestClassifier(random_state=42,n_estimators=700)
RF.fit(xtrain_final, ytrain)

ypred3=RF.predict(xtest_final);
```

```python
print("Random Forest Classifier accuracy: ", RF.score(xtest_final,ytest))
print(confusion_matrix(ytest, ypred3))
print("Random Forest Classifier f1 score: ", f1_score(ytest, ypred3, average="macro"))
```
