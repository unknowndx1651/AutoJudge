### Regression

```python
# import os
# os.chdir("..")

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

from src.regressor import xtrain_final,xtest_final,ytrain,ytest
```

### 1.Linear Regression

```python
LR=LinearRegression()
LR.fit(xtrain_final, ytrain)

ypred1=LR.predict(xtest_final);
```

```python
print("Linear Regression accuracy: ", LR.score(xtest_final,ytest))
print("Linear Regression MAE: ", mean_absolute_error(ytest,ypred1))
print("Linear Regression RMSE: ", root_mean_squared_error(ytest, ypred1))
```

### 2. Gradient Boosting Regressor

```python
GBR=GradientBoostingRegressor(n_estimators=300,learning_rate=0.1,random_state=42)
GBR.fit(xtrain_final, ytrain)

ypred2=GBR.predict(xtest_final);
```

```python
print("Gradient Boosting accuracy: ", GBR.score(xtest_final,ytest))
print("Gradient Boosting MAE: ", mean_absolute_error(ytest,ypred2))
print("Gradient Boosting RMSE: ", root_mean_squared_error(ytest, ypred2))
```

### 3. Random Forest Regressor:

```python
RF=RandomForestRegressor(random_state=42,n_estimators=100)
RF.fit(xtrain_final, ytrain)

ypred3=RF.predict(xtest_final);
```

```python
print("Random Forest Classifier accuracy: ", RF.score(xtest_final,ytest))
print("Random Forest MAE: ", mean_absolute_error(ytest,ypred3))
print("Random Forest RMSE: ", root_mean_squared_error(ytest, ypred3))
```
